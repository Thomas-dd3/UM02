```python
import pandas as pd
```

```python
# Matplotlib for additional customization
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Seaborn for plotting and styling
import seaborn as sns
```

```python
df = pd.read_csv('UTSEUS-anjuke-real-estate.csv')
```

```python
df.head()
```

```python
sns.set()
```

```python
sns.lmplot(x='longitude', y='latitude', data=df)
```

```python
sns.lmplot(x='surface', y='price', data=df)
plt.xscale('log')
plt.yscale('log')
```

```python
sns.lmplot(x='surface', y='price', data=df,
           fit_reg=False, # No regression line
           hue='district')
plt.xscale('log')
plt.yscale('log')
```

```python
plt.figure(figsize=(12,9))
df2 = df.drop(columns = ['price','onesquaremeter', 'surface', 'longitude', 'latitude', 'done'])
sns.boxplot(data=df2)
```

```python
plt.figure(figsize=(12,9))
sns.boxplot(data=df['price'])
```

```python
# Set theme
sns.set_style('whitegrid')
plt.figure(figsize=(20,12))
 
# Violin plot
sns.violinplot(x='district', y='onesquaremeter', data=df)
plt.yscale('log')
```

```python
plt.figure(figsize=(14,10))
# Density Plot
sns.kdeplot(df.longitude, df.latitude)
```
